```python
import math
f=open('Homework 4 Output.txt','w')
for t in range(19,30):
    def UTcon(t,d,m,y,TZ = +5):
        """
        t: int - Current time in the number of whole hours since midnight
        TZ: int - The time zone of the location in question in hours
        Returns UT: - The Current time accurate to the hour in Universal Time
        """
        UT=t+TZ
        if(UT>=24):
            UT=UT-24
            d=d+1
        if(((m==1) or (m==3) or (m==5) or (m==7) or (m==8) or (m==10)) and (d>31)):
            m=m+1
            d=1
        if(((m==4) or (m==6) or (m==9) or (m==11)) and (d>30)):
            m=m+1
            d=1
        if((m==2) and ((int(y/4)-(y/4.0))==0) and (d>29)):
            m=m+1
            d=1
        if((m==2) and (((int(y/4))-(y/4.0))!=0) and (d>28)):
            m=m+1
            d=1
        if((m==12) and (d>31)):
            m=1
            d=1
            y=y+1
        return UT,d,m,y
    UT, Ud, Um, Uy = UTcon(t,1,4,2016,7)
    f.write("UT ; "+str(UT)+" hour")
    f.write(" "+str(Ud)+" day")
    f.write(" "+str(Um)+" month")
    f.write(" "+str(Uy)+" year"+"\n")
    def LSTcon(UT,Ud,Um,Uy,Wlon):
        """
        UT: float - an input Universal Time
        Ud: int - an input for the day in UT
        Um: int - an input for the month in UT
        Uy: int -an input for the year in UT
        Wlon: float- an input for the longitude measured westward from Greenwich in decimal hours
        Returns LST for the given loctation and time
        """
        if(Um==1):
            GMST0=6.6070
        if(Um==2):
            GMST0=8.6440
        if(Um==3):
            GMST0=10.5496
        if(Um==4):
            GMST0=12.5866
        GMST=GMST0+(0.065710*(Ud))+(1.002738*(UT))
        LST=GMST-Wlon
        return LST
    LST=LSTcon(UT,Ud,Um,Uy,7.4398)
    f.write("LST ; "+str(LST)+"h"+"\n")
    def HAObj(LST,RA):
        """
        LST: float- an input Local Sidereal Time
        RA: float- the Right Ascension of an object in decimal hours
        Returns- The hour angle of the given object at the given time

        This function ouputs the hour angle of an object given the local sidereal time and the right ascension
        """
        HA=LST-RA
        return HA
    HA=HAObj(LST,12.44366667)
    f.write("HA ; "+str(HA)+"h"+"\n")
    def AltObj(HA,Lat,Dec):
        """
        HA: float - The hour angle of the given object at the given time in decimal hours
        Lat: float - The latitde of the observering location in decimal degrees
        Dec: float - The declination of an object in decimal degrees
        Returns - The Altitude of an object in decimal degrees

        This function finds the altiude of an object given the
        hour angle of the object, the latitude of the observing location and the declination of the object
        """
        Lat=Lat*((math.pi)/180)
        Dec=Dec*((math.pi)/180)
        HA=HA*((math.pi)/12)
        Alt= math.asin(((math.sin(Lat))*(math.sin(Dec)))+((math.cos(Lat))*(math.cos(Dec))*(math.cos(-HA))))
        Alt=Alt*(180/(math.pi))
        return Alt
    Alt=AltObj(HA,31.96,33.4569)
    f.write("Alt ; "+str(Alt)+" degrees"+"\n")
    def AirmassObj(Alt):
        """
        Alt: float - The Altitude of an object in decimal degrees
        Retruns: float - The Airmass beween the observer and the object

        This function calculates that airmass of a given object given its altitude in decimal degrees as an input
        """
        Alt=Alt*((math.pi)/180)
        Airmass=(1/(math.cos(((math.pi)/2)-Alt)))
        return Airmass
    Airmass=AirmassObj(Alt)
    f.write("Airmass ; "+str(Airmass)+"\n")
    f.write("Next Hour" +"\n"+"\n")
f.close()
```
